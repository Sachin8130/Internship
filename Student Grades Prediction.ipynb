```python
col_null=df.columns[df.isnull().any()].tolist()
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(strategy="most_frequent")
df[col_null]=pd.DataFrame(si.fit_transform(df[col_null]),columns=df[col_null].columns)
```

```python
df.isnull().sum().sum()
```

```python
for i in col_null:
    print(i,df[i].unique())
```

```python
df.drop('Seat No.',axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in df.columns:
    if df[i].dtypes=='object':
        df[i]=le.fit_transform(df[i])
```

```python
df
```

```python
x=df.drop('CGPA',axis=1)
```

```python
y=df['CGPA']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

```python
MaxAcc=0
MaxRS=0
for i in range(0,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test,)
    acc=r2_score(y_test,pred)
    if acc>MaxAcc:
        MaxAcc=acc
        MaxRS=i
print("Maximum R2 Score=",MaxAcc,"Maximum Random State=",MaxRS)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=MaxRS)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Lasso,Ridge
from sklearn.svm import SVR
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
predLR=lr.predict(x_test)
predLR_train=lr.predict(x_train)
print("r2 score:",r2_score(y_test,predLR))
print("r2 score on train data:",r2_score(y_train,predLR_train))
print("Mean Absolute Error:",mean_absolute_error(y_test,predLR))
print("Mean_square_error",mean_squared_error(y_test,predLR))
print("RMSE:",np.sqrt(mean_squared_error(y_test,predLR)))
```

```python
GBR=GradientBoostingRegressor()
GBR.fit(x_train,y_train)
predGBR=GBR.predict(x_test)
predGBR_train=GBR.predict(x_train)
print("R2 Score:",r2_score(y_test,predGBR))
print("R2 score on train data=",r2_score(y_train,predGBR_train))
print("Mean Absolute Error:",mean_absolute_error(y_test,predGBR))
print("Mean Squared Error:",mean_squared_error(y_test,predGBR))
print("RMSE:",np.sqrt(mean_squared_error(y_test,predGBR)))
```

```python
DTR=DecisionTreeRegressor()
DTR.fit(x_train,y_train)
predDTR=DTR.predict(x_test)
predDTR_train=DTR.predict(x_train)
print("R2 Score:",r2_score(y_test,predDTR))
print("R2 score on train data=",r2_score(y_train,predDTR_train))
print("Mean Absolute Error:",mean_absolute_error(y_test,predDTR))
print("Mean Squared Error:",mean_squared_error(y_test,predDTR))
print("RMSE:",np.sqrt(mean_squared_error(y_test,predDTR)))
```

```python
LA=Lasso()
LA.fit(x_train,y_train)
predLA=LA.predict(x_test)
predLA_train=LA.predict(x_train)
print("R2 Score:",r2_score(y_test,predLA))
print("R2 score on train data=",r2_score(y_train,predLA_train))
print("Mean Absolute Error:",mean_absolute_error(y_test,predLA))
print("Mean Squared Error:",mean_squared_error(y_test,predLA))
print("RMSE:",np.sqrt(mean_squared_error(y_test,predLA)))
```

```python
RD=Ridge()
RD.fit(x_train,y_train)
predRD=RD.predict(x_test)
predRD_train=RD.predict(x_train)
print("r2 Score:",r2_score(y_test,predRD))
print("r2 score on train data:",r2_score(y_train,predRD_train))
print("MAE:",mean_absolute_error(y_test,predRD))
print("MSE:",mean_squared_error(y_test,predRD))
print("RMSE:",np.sqrt(mean_squared_error(y_test,predRD)))
```

```python
knn=KNN()
knn.fit(x_train,y_train)
predknn=knn.predict(x_test)
predknn_train=knn.predict(x_train)
print("R2 Score:",r2_score(y_test,predknn))
print("R2 score on train data=",r2_score(y_train,predknn_train))
print("Mean Absolute Error:",mean_absolute_error(y_test,predknn))
print("Mean Squared Error:",mean_squared_error(y_test,predknn))
print("RMSE:",np.sqrt(mean_squared_error(y_test,predknn)))
```

```python
svr=SVR()
svr.fit(x_train,y_train)
predsvr=svr.predict(x_test)
predsvr_train=svr.predict(x_train)
print("R2 Score:",r2_score(y_test,predsvr))
print("R2 score on train data=",r2_score(y_train,predsvr_train))
print("Mean Absolute Error:",mean_absolute_error(y_test,predsvr))
print("Mean Squared Error:",mean_squared_error(y_test,predsvr))
print("RMSE:",np.sqrt(mean_squared_error(y_test,predsvr)))
```
