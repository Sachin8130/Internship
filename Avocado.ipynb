```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve
```

```python
MaxAcc=0
MaxRS=0
for i in range(1,20):
    x1_train,x1_test,y1_train,y1_test=train_test_split(x2,y2,test_size=.30,random_state=i)
    LR=LogisticRegression()
    LR.fit(x1_train,y1_train)
    pred=LR.predict(x1_test)
    acc=accuracy_score(y1_test,pred)
    if acc>MaxAcc:
        MaxAcc=acc
        MaxRS=i
print("Max Accuracy:",MaxAcc,"MaxRS:",MaxRS)
```
